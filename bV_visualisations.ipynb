```python
import math

import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

# **Bianchi V visualisations**

## Omega2 = 1e-3

```python
# sns.set(rc={'axes.facecolor': 'white'})
```

```python
df_bV_upside = pd.read_csv('bianchiV_isotropic_eta_0.018000_omega_2_0.001000_1000p_upside.txt', sep=" ", header=None)
df_bV_upside.columns = ['x', 'a', 'da', 'phi']
df_bV_upside['da/a'] = df_bV_upside['da'] / df_bV_upside['a']
df_bV_upside['x'] = df_bV_upside['x'] + 2.36
```

```python
sns.lineplot(df_bV_upside, x='x', y='da/a')
plt.ylim(0, 5)
plt.xlim(0, 8)
```

## Omega2 = 1e-5

```python
df_bV_upside = pd.read_csv('bianchiV_isotropic_eta_0.018000_omega_2_0.000010_1000p_upside.txt', sep=" ", header=None)
df_bV_upside.columns = ['x', 'a', 'da', 'phi']
df_bV_upside['da/a'] = df_bV_upside['da'] / df_bV_upside['a']
df_bV_upside['x'] = df_bV_upside['x'] + 3.28
```

```python
sns.lineplot(df_bV_upside, x='x', y='da/a')
plt.ylim(0, 5)
plt.xlim(0, 8)
```

## Omega2 = 1e-5, s=0.78

```python
df_bV_upside = pd.read_csv('bianchiV_anisotropic_eta_0.018000_omega_2_0.000010_s_0.780000_1000p_upside.txt', sep=" ", header=None)
df_bV_upside.columns = ['x', 'a', 'da', 'phi', 's']
df_bV_upside['da/a'] = df_bV_upside['da'] / df_bV_upside['a']
df_bV_upside['x'] = df_bV_upside['x'] + 3.1
```

```python
sns.lineplot(df_bV_upside, x='x', y='da/a')
sns.lineplot(df_bV_upside, x='x', y='s')
plt.ylim(0, 5)
plt.xlim(0, 8)
```
