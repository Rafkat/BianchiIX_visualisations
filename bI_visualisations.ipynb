```python
import math

import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df_bI_upside_from_bIX = pd.read_csv('bianchiI_isotropic_eta_0.018000_omega_2_0.000000_1000p_upside_from_bIX.txt', sep=" ", header=None)
df_bI_upside_from_bIX.columns = ['x', 'a', 'da', 'phi']
df_bI_upside_from_bIX['da/a'] = df_bI_upside_from_bIX['da'] / df_bI_upside_from_bIX['a']
df_bI_upside_from_bIX['x'] = df_bI_upside_from_bIX['x'] + 3.2878
df_bI_upside_from_bIX
```

```python
sns.lineplot(df_bI_upside_from_bIX, x='x', y='da/a')
```

```python
df_bI_upside_from_bV = pd.read_csv('bianchiI_isotropic_eta_0.018000_omega_2_0.000000_1000p_upside_from_bV.txt', sep=" ", header=None)
df_bI_upside_from_bV.columns = ['x', 'a', 'da', 'phi']
df_bI_upside_from_bV['da/a'] = df_bI_upside_from_bV['da'] / df_bI_upside_from_bV['a']
df_bI_upside_from_bV['x'] = df_bI_upside_from_bV['x'] + 3.2878
df_bI_upside_from_bV
```

```python
sns.lineplot(df_bI_upside_from_bV, x='x', y='da/a')
```

```python
fig, ax = plt.subplots()
sns.lineplot(df_bI_upside_from_bIX, x='x', y='da/a')
sns.lineplot(df_bI_upside_from_bV, x='x', y='da/a')
```

```python
df_bI_upside_a_from_bV = pd.read_csv('bianchiI_anisotropic_eta_0.018000_omega_2_0.000000_s_0.000000_1000p_upside_from_bV.txt', sep=" ", header=None)
df_bI_upside_a_from_bV.columns = ['x', 'a', 'da', 'phi', 's']
df_bI_upside_a_from_bV['da/a'] = df_bI_upside_a_from_bV['da'] / df_bI_upside_a_from_bV['a']
df_bI_upside_a_from_bV['x'] = df_bI_upside_a_from_bV['x'] + 3.2896
```

```python
sns.lineplot(df_bI_upside_a_from_bV, x='x', y='da/a')
sns.lineplot(df_bI_upside_a_from_bV, x='x', y='s')
plt.ylim(0, 3)
plt.xlim(0, 8)
```

```python
sns.lineplot(df_bI_upside_a_from_bV, x='x', y='s')
```

```python
df_bI_upside_a_from_bIX = pd.read_csv('bianchiI_anisotropic_eta_0.018000_omega_2_0.000000_dbp_0.000000_dbm_0.000000_bp_0.000000_bm_0.000000_1000p_upside_from_bIX.txt', sep=" ", header=None)
df_bI_upside_a_from_bIX.columns = ['x', 'a', 'da', 'phi', 'dbp', 'dbm', 'bp', 'bm']
df_bI_upside_a_from_bIX['da/a'] = df_bI_upside_a_from_bIX['da'] / df_bI_upside_a_from_bIX['a']
df_bI_upside_a_from_bIX['x'] = df_bI_upside_a_from_bIX['x'] + 3.42122
df_bI_upside_a_from_bIX
```

```python
sns.lineplot(df_bI_upside_a_from_bIX, x='x', y='da/a')
plt.ylim(0, 5)
plt.xlim(0, 20)
```

```python
df_bI_upside_a_from_bIX['s'] = (df_bI_upside_a_from_bIX['dbp'].apply(lambda x: x ** 2) + df_bI_upside_a_from_bIX['dbm'].apply(lambda x: x ** 2)).apply(lambda x: math.sqrt(x))
```

```python
sns.lineplot(df_bI_upside_a_from_bIX, x='x', y='s')
```

```python
fig, ax = plt.subplots()
sns.lineplot(df_bI_upside_a_from_bIX, x='x', y='s')
sns.lineplot(df_bI_upside_a_from_bV, x='x', y='s')
```
